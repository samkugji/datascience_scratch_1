# 로지스틱 회귀
- $lnL=\sum  y_{ i }ln\{ \theta (w^{ T }x_{ i })^{ y_{ i } }\} +\sum  (1-y_{ i })ln\{ 1-\theta (w^{ T }x_{ i })\} ^{ 1-y_{ i } }$
    - 로지스틱 함수
        - ### $\theta(x;w)=\frac{1}{1+exp(-w^Tx)}$
- 로그는 단조 증가함수이므로 log-likelihood를 최대로하는 w는 동시에 likelihood를 최대화한다.
- 로지스틱 회귀식은 w에 재한 비선형 함수이므로 OLS처럼 한번에 풀리지 않는다. 따라서 수치적 최적화를 실시한다.

```python
import statsmodels.api as sm
```

```python
logit_mod = sm.Logit(Y, X)
logit_res = logit_mod.fit(disp=0)
print(logit_res.summary())
```

```python
## 시각화
plt.figure(figsize = (10, 5))
plt.xlabel('predicted probability')
plt.ylabel('actual outcome')
plt.scatter(logit_res.predict(X), Y, label = 'real', alpha = 0.8)
plt.legend()
```

```python
## 성능 평가

true_positives = false_positives = true_negatives = false_negatives = 0

for x_i, y_i in zip(X, Y):
    predict = logit_res.predict(x_i)
    if y_i == 1 and predict >= 0.5:  # TP: paid and we predict paid
        true_positives += 1
    elif y_i == 1:                   # FN: paid and we predict unpaid
        false_negatives += 1
    elif predict >= 0.5:             # FP: unpaid and we predict paid
        false_positives += 1
    else:                            # TN: unpaid and we predict unpaid
        true_negatives += 1

precision = true_positives / (true_positives + false_positives)
recall = true_positives / (true_positives + false_negatives)
```

```python
print('precision : ' , precision)
print('유료계정 유저인데 유료계정 유저라고 맞춘 수 / (유료계정 유저인데 유료계정 유저라고 맞춘 수 + 유료계정 유저인데 유료계정 유저가 아니라고 맞춘 수)')
```

```python
print('recall : ' , recall)
print('유료계정 유저인데 유료계정 유저라고 맞춘 수 / (유료계정 유저인데 유료계정 유저라고 맞춘 수 + 유료계정 유저가 아닌데 유료계정 유저라고 맞춘 수)')
```
